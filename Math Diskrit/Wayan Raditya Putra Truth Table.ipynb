# No 1 a

```python
import pandas as pd

dfs = []
for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            prop1 = not q
            prop2 = not q or r
            prop3 = prop2 if p else True #jika p benar maka prop 3 nilainya sama dengan prop2, kalau p salah prop3 pasti bernilai bemar
            dfs.append(pd.DataFrame({'P': [p],
                                    'Q': [q],
                                    'R' : [r],
                                    '~Q':[prop1],
                                    '~Q V R': [prop2],
                                    'P ==> (~Q V R)' : [prop3]
                                    }))

tb_soal = pd.concat(dfs, ignore_index=True)
tb_soal
```

# No 2 (Mengecek Tautologi)

Show that each of these conditional statements is a tautology by using truth tables.
a) [¬p ∧ (p ∨ q)] → q
b) [(p → q) ∧ (q → r)] → (p → r)
c) [p ∧ (p → q)] → q
d) [(p ∨ q) ∧ (p → r) ∧ (q → r)] → r

```python
import pandas as pd

dfs = []

for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            # ~p
            prop1 = not p                      
            # ~q
            prop2 = not q                      
            # ~r
            prop3 = not r                      
            # p ∨ q
            prop4 = p or q                     
            # p → q
            prop5 = q if p else True           
            # p → r
            prop6 = r if p else True           
            # q → r
            prop7 = r if q else True           
            # ~p ∧ (p ∨ q)
            prop8 = not p and prop4            
            # (p → q) ∧ (q → r)
            prop9 = prop5 and prop7            
            # p ∧ (p → q)
            prop10 = p and prop5               
            # (p ∨ q) ∧ (p → r) ∧ (q → r)
            prop11 = prop4 and prop6 and prop7 
            # [~p ∧ (p ∨ q)] → q
            prop12 = q if prop8 else True
            # [(p → q) ∧ (q → r)] → (p → r)
            prop13 = prop6 if prop9 else True
            # [p ∧ (p → q)] → q
            prop14 = q if prop10 else True
            # [(p ∨ q) ∧ (p → r) ∧ (q → r)] → r
            prop15 = r if prop11 else True
        
            dfs.append(pd.DataFrame({
                'P': [p],
                'Q': [q],
                '-P': [prop1],
                '-Q': [prop2],
                '-R': [prop3],
                'p ∨ q': [prop4],
                'p → q': [prop5],
                'p → r': [prop6],
                'q → r': [prop7],
                '~p ∧ (p ∨ q)': [prop8],
                '(p → q) ∧ (q → r)': [prop9],
                'p ∧ (p → q)': [prop10],
                '(p ∨ q) ∧ (p → r) ∧ (q → r)': [prop11],
                '[~p ∧ (p ∨ q)] → q': [prop12],
                '[(p → q) ∧ (q → r)] → (p → r)': [prop13],
                '[p ∧ (p → q)] → q': [prop14],
                '[(p ∨ q) ∧ (p → r) ∧ (q → r)] → r': [prop15],
            }))

tb_soal = pd.concat(dfs, ignore_index=True)
tb_soal
```

Show that (p ∧ q) → r and (p → r) ∧ (q → r) are not logically equivalent.

# No 4 (Mengecek Siapa Yang Bohong Gaes)

Seorang detektif sedang menangani sebuah kasus kriminal. Ada 4 orang tersangka berinisial G, O, L, D. Dari hasil interogasi, detektif tersebut mendapatkan sejumlah fakta, yaitu :
Jika G berkata jujur, maka O berkata jujur.
Jika L berkata bohong, maka D berkata jujur.
Salah satu dari O dan L berkata jujur.
O berkata jujur jika dan hanya jika D berkata bohong. Dari sejumlah fakta di atas, siapa yang berkata jujur? buatlah pembuktiannya dengan tabel kebenaran!

```python
import pandas as pd
dfs = []
for g in [True, False]:
    for o in [True, False]:
        for l in [True, False]:
            for d in [True, False]:
                prop1 = not g
                prop2 = not o
                prop3 = not l
                prop4 = not d
                

                #prop kesaksian dari 4 tersangka
                prop5 = o if g else True
                prop6 = d if not l else True
                prop7 = bool(o != l)
                prop8 = bool(prop4 == o)
                dfs.append(pd.DataFrame({'G': [g],
                                         'O': [o],
                                         'L': [l],
                                         'D': [d],
                                         '-G': [prop1],
                                         '-O': [prop2],
                                         '-L': [prop3],
                                         '-D': [prop4],
                                         'G → O': [prop5],
                                         '-L → D': [prop6],
                                         'O xor L': [prop7],
                                         '-D ↔ O': [prop8],
                                        }))
tb_soal = pd.concat(dfs, ignore_index=True)
```

```python
#Jika G Jujur
tb_soal[tb_soal["G"] == True]
```

```python
#Jika O Jujur
tb_soal[tb_soal["O"] == True]
```

```python
#Jika L jujur
tb_soal[tb_soal["L"] == True]
```
